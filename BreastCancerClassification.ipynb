## Breast Cancer Classification with Neural Networks

- 1 - benign
- 0 - malignant

```python
#importing dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets # has many datasets to load
from sklearn.model_selection import train_test_split
```

```python
#Data collection and processing
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
print(breast_cancer_dataset)
```

```python
#loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
# adding the target column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# Seperating the features and target
X = data_frame.drop(columns = 'label', axis = 1) # axis = 0 if row dropped
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Standardize the data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler();
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

```python
#Building the neural network
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# Setting the layers of Neural Network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation= 'relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# Compiling the neural network
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#training the neural network
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10); # okay i messed up by not saving the states for 10 epochs and then going ahead for 100 epochs
```

```python
# visualizing accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc =  'lower right')
```

```python
# visualizing loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc =  'upper right')
```

```python
# Accuracy of the model on test data
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
Y_pred = model.predict(X_test_std)
print(Y_pred.shape)
print(Y_pred[0]) # probability of malignant is 66.43% and benign is 96.4%
```
